```python
import pandas as pd
df=pd.read_csv('Amazon.csv')
print("Available columns:")
print(df.columns.tolist())
```

```python
df.columns.tolist()
```

```python
total_sales=df["TotalAmount"].sum()
print("Total Sales:",total_sales)
```

```python
total_orders= df["OrderID"].nunique()
print("Total Orders:",total_orders)
```

```python
average_sales=df["TotalAmount"].mean()
print("Average Sales:",average_sales)
```

```python
max_sales=df["TotalAmount"].min()
print("Maximum Sales:", max_sales)
```

```python
min_sales=df["TotalAmount"].min()
print("Minimum Sales:",min_sales)
```

```python
sales_by_state = df.groupby("State")["TotalAmount"].sum().sort_values(ascending=False)
print(sales_by_state)
```

```python
sales_by_category = df.groupby("Category")["TotalAmount"].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
top5_customers = df.groupby("CustomerName")["TotalAmount"].sum().sort_values(ascending=False).head(5)
print(top5_customers)
```

```python
top5_products = df.groupby("ProductName")["TotalAmount"].sum().sort_values(ascending=False).head(5)
print(top5_products)
```

```python
payment_method= df.groupby("PaymentMethod")["TotalAmount"].sum().sort_values(ascending=False).head(5)
print(payment_method)
```

```python
df["OrderDate"]=pd.to_datetime(df["OrderDate"])
```

```python
df["Month"]=df["OrderDate"].dt.month_name()
```

```python
monthly_sales = df.groupby("Month")["TotalAmount"].sum()
print(monthly_sales)
```

```python
sales_by_state.plot(kind="bar", figsize=(10,5))
plt.title("Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_category.plot(kind="pie", autopct="%1.1f%%", figsize=(7,7))
plt.title("Sales by Category")
plt.ylabel("")
plt.show()
```

```python
monthly_sales.plot(kind="line", marker="o", figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
top10_products = df.groupby("ProductName")["TotalAmount"].sum().sort_values(ascending=False).head(10)
```

```python
top10_products.plot(kind="barh", figsize=(10,6))
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product")
plt.show()
```
